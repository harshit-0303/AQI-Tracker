```python
import pandas as pd
```

```python
cities = pd.read_csv("cities1000.txt", sep="\t", header=None)
states = pd.read_csv("admin1CodesASCII.txt", sep="\t", header=None)
```

```python
cities
```

```python
states
```

```python
india = cities[cities[8] == "IN"]

india["state_code"] = "IN." + india[10].astype(str)

merged = india.merge(states, left_on="state_code", right_on=0)

```

```python
merged.head(2)
```

```python
merged['state'] = merged['1_y']
merged['city'] = merged['1_x']
merged['latitude'] = merged[4]
merged['longitude'] = merged[5]
df = merged[['state_code','state','city','latitude','longitude']]
```

```python
df['state'].value_counts()
```

```python
df.to_csv("State_geo_list.csv",index = False)
```
